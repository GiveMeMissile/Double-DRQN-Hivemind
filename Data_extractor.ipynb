DOCUMENTATION (Graph functions only):
  
graph() function:  
graph(x, data, names, x_name): This function graphs multiple different lines of data vs an x value.  
x (list): A list which contains all of the x points which will be graphed  
data (2d list): a 2d list which contains all the data which will be graphed, the list contains multiple lists of data what will be graphed vs the x data.  
names (list of strings): A list of all the names for the data held within the data parameter.  
x_name (strings): The name of the x data.  
  
graph_scatter() function:  
graph_scatter(x, y, x_name, y_name): This function graphs a scatterplot with all of the inputted points given.  
x (list): A list which contains all of the x points which will be graphed  
y (list): A list which contains all of the y points which will be graphed  
x_name (string): The name of the x data.  
y_name (string): The name of the y data.  
  
graph_lsrl() function:  
graph_lsrl(self, x, y, x_name, y_name, scatter=False): This function calculates and graphs the lsrl of the given data. And it can also graph the scatter plot as well.  
x (list): A list which contains all of the x points which will be graphed  
y (list): A list which contains all of the y points which will be graphed  
x_name (string): The name of the x data.  
y_name (string): The name of the y data.  
scatter (boolean): Boolean value which when true, shows the scatter plot along with the lsrl.  
  
graph_box_plot() function:  
graph_box_plot(self, data, name): Function which graphs a box plot based off of the inputted data.  
data (list): A list which contains all of the data which will be used to create the box plot.  
name (string): The name of the data which is being used to create the boxplot.  
  
graph_histogram() function:  
graph_histogram(self, data, name, bins=10): This function creates a Histogram based off of the given data.  
data (list): A list which contains all of the data which will be used to create the Histogram.  
name (string): The name of the data which is being used to create the Histogram.  
bins (int): The number or bars which the Histogram will have.

```python
x_name = "Episodes"
y_names = ["Q Values Average", "Q Values Max", "Q Values Min", "Rewards"]# , "Actions", "Epsilon", "Rewards", "Health"]
y_data = []

for piece in data:
    specific_data = []
    for name in y_names:
        specific_data.append(piece[name])
    y_data.append(specific_data)
    specific_data = []

for piece, y_piece in zip(data, y_data):
    data_analyzer.graph(piece[x_name], y_piece, y_names, x_name)
```

```python
name = "Q Values Average"

for piece in data:
    data_analyzer.graph_box_plot(data=piece[name], name=name)
```

```python
name = "Q Values Max"

for piece in data:
    data_analyzer.graph_histogram(piece[name], name, bins=10)
```

```python
x_name = "Episodes"
y_name = "Rewards"
scatter_display = True

for piece in data:
    data_analyzer.graph_lsrl(piece[x_name], piece[y_name], x_name, y_name, scatter=scatter_display)
```
